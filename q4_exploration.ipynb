## Part 4: Value Counts and Grouping (5 points)

1. Get value counts for the 'site' column
2. Get value counts for the 'intervention_group' column  
3. Create a crosstab of site vs intervention_group
4. Calculate mean age by site
5. Save the site value counts to `output/q4_site_counts.csv`

```python
# TODO: Value counts and analysis
df['site'] = df['site'].str.strip().str.lower()
df['site'] = df['site'] .replace({"site_d": "site d", "site  a": "site a"})
# print(np.unique(df['site']))
df['intervention_group'] = df['intervention_group'].str.strip().str.lower()
df['intervention_group'] = df['intervention_group'].replace({"contrl": "control", "treatment  b": "treatment b", "treatmeanta": "treatment a", "treatmenta": "treatment a", "treatmen a": "treatment a"})
# print(np.unique(df['intervention_group']))
site_counts = df["site"].value_counts()
intervention_group_counts = df["intervention_group"].value_counts()
cross_tab = pd.crosstab(df["site"], df["intervention_group"])
mean_age_by_site = df.groupby("site")["age"].mean()

print(f"Value counts for 'site' column: {site_counts}")
print(f"Value counts for 'intervention_group' column: {intervention_group_counts}")
print(f"Cross tab of site vs intervention_group {cross_tab}")
```
